```python
a = [16,18,23,12,17,14,7,0,23,11,21,19,22,15,7,14,9,12,11,10,26,19,11,14,4,12,22,18,17,7,20,23,23,18,24,23,19,19,11,15,22,17,17,17,18,16,3,13,2,11,22,21,10,27,8,17,15,15,15]
dict = {}
for i in a:
    if 0 <= i <= 4:
        dict['0 - 4'] = dict.get('0 - 4', 0) + 1
    elif 5 <= i <= 9:
        dict['5 - 9'] = dict.get('5 - 9', 0) + 1
    elif 10 <= i <= 14:
        dict['10 - 14'] = dict.get('10 - 14', 0) + 1
    elif 15 <= i <= 19:
        dict['15 - 19'] = dict.get('15 - 19', 0) + 1
    elif 20 <= i <= 24:
        dict['20 - 24'] = dict.get('20 - 24', 0) + 1
    elif 25 <= i <= 30:
        dict['25 - 30'] = dict.get('25 - 30', 0) + 1

for key, value in dict.items():
    print(f"{key} --> {value}")

print(f"Total students who appeared in exam: {sum(dict.values())}")
```

```python
import matplotlib.pyplot as plot
x = list(dict.keys())
y = list(dict.values())
plot.bar(x, y)
plot.title("Sec 28 Marks")
plot.xlabel("Mark range")
plot.ylabel("No. of students")
```
